```python
def directory(s):
    files = os.listdir(s)
    return files, os.getcwd()
```

```python
def open_csv(s):
    files = os.listdir(s)
    for file in files:
        if ".csv" in file:    
            curr_dir = os.path.join(s,file)
            data = pd.read_csv(curr_dir,encoding='latin')
            
        else:
            print( s + " Does not contain any csv file format")
    return data
```

```python
netflix_df = open_csv('.')
```
